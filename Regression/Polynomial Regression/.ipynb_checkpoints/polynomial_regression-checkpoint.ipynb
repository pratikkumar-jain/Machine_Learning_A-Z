```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#importing the dataset
dataset = pd.read_csv("Position_Salaries.csv")
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
lin_reg1 = LinearRegression()
lin_reg1.fit(X, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2) #degree 2
X_poly = poly_reg.fit_transform(X)
```

```python
X_poly
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly, y)
```

```python
#visualization of Linear regression
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg1.predict(X), color = 'blue')
plt.title("Graph for Linear Regression")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
#visualization of Polynomial regression
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title("Graph for Polynomial Regression")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
poly_reg = PolynomialFeatures(degree=3) #degree 3
X_poly = poly_reg.fit_transform(X)
```

```python
lin_reg3 = LinearRegression()
lin_reg3.fit(X_poly, y)
```

```python
#visualization of Polynomial regression
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg3.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title("Graph for Polynomial Regression")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
poly_reg = PolynomialFeatures(degree=4) #degree 4
X_poly = poly_reg.fit_transform(X)

lin_reg4 = LinearRegression()
lin_reg4.fit(X_poly, y)

#visualization of Polynomial regression
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg4.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title("Graph for Polynomial Regression")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
#visualization of Polynomial regression
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg4.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title("Graph for Polynomial Regression")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
#prediction
lin_reg1.predict(6.5)
```

```python
lin_reg4.predict(poly_reg.fit_transform(6.5))
```
